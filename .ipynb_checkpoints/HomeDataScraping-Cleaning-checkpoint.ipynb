#
# Imports

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
```

#
# Reading CSV files with all the links

```python
links = pd.read_csv('HomeLinks.csv')
links
```

#
# Scraping

###
### Converting the links DataFrame into a list to iterate through

```python
links = links.values.tolist()
```

```python
len(links)
```

```python
links
```

###
### We will now iterate through each webpage, and extract the data we want

```python
final = []
i=0
for link in links:
    

    try:
        data = {}
    
        temp = []
        
        url = f'https://www.weichert.com{link[0]}'
        res = requests.get(url)
        soup = BeautifulSoup(res.text, 'html.parser')
    
        # Scraping Market Price and Sq Ft
        mainDiv = soup.find('div', class_='col-sm-9 col-md-9 propertyAddressSection')
    
        for element in mainDiv:
            temp.append(element.text.strip())
            
        temp.pop(0)
        price = float(temp[1].replace('$', '').replace(',','').strip())
        sqft = int(temp[-1].split()[-1].replace(',',''))
        data = {'Price' : price, 'Sq Ft' : sqft}
    
        # Scraping other information (varies)
        mainDiv = soup.find('div', class_='listing-features shift-div-left')
        for primary in mainDiv:
            for secondary in primary:
                divs = secondary.find_all('div', class_='col-xs-6')
                if divs[0].text != 'Price':
                    data[divs[0].text] = divs[-1].text

        # Average List Price and Median Income of town/city (Cleaned on the spot)
        mainDiv = soup.find('div', class_='property-type-stat')
        listPrice = mainDiv.find('span', class_='market-list-value').text.replace(',','').replace('$','')
        mainDiv = soup.find('div', id='medianIncome')
        medianIncome = mainDiv.find('span', class_='population-data-display').text.replace(',','').replace('$','').strip()
        data['Town/City Median Income'] = medianIncome
        final.append(data)
        i+=1
        print(f'Iteration - {i}')

    
    except:
        print('Error. Link will be ignored')
        

    
```

```python
len(final)
```

```python
final = pd.DataFrame(final)
```

#
# Cleaning/Wrangling

###
### Changing all null values for column: 'Half bath', to 0
### Also removing all rows with a null value. This is done afterwards so we dont drop homes with 0 half-baths

```python
final.loc[final['Half Bath'].isnull(), 'Half Bath']= 0
final = final.dropna()
final = final.reset_index(drop=True)
final.head(3)
```

###
### Removing all Studios, Condos, TownHouses, and Co-Ops

```python
final = final[final['Bedrooms'] != ' Studio ']
final = final[final['Property Type'] != ' Condo ']
final = final[final['Property Type'] != ' Townhouse ']
final = final[final['Property Type'] != ' Co-Op ']


final = final.reset_index(drop=True)
```

###
### Dropping unnecessary columns

```python
final.drop(['MLS | Web ID'], axis=1, inplace=True)
final.drop(['County/Region'], axis=1, inplace=True)
final.drop(['Style'], axis=1, inplace=True)
final.drop(['Taxes'], axis=1, inplace=True)
final.head()
```

```python
final['Town/City'].value_counts()
```

###
# FIRST/SECOND BATCH CODE IGNORE
### If we take a look at the value counts of the column: Town/City, there are two distinct values: Garden City South and Stewart Manor
### We will now change these values to Garden City since the zip codes are the same despite the different value

```python
outsider = final['Town/City'].value_counts().index[-1]
index = final[final['Town/City'] == outsider].index[0]
final.loc[index, 'Town/City'] = 'Garden City'

final['Town/City'].value_counts()
```

```python
outsider = final['Town/City Median Income'].value_counts().index[-1]
index = final[final['Town/City Median Income'] == outsider].index
final.loc[index, 'Town/City Median Income'] =  148988

final['Town/City Median Income'].value_counts()
```
